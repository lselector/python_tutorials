### Three-Dimensional plot examples

```python
import numpy as np

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm  # color maps

# choose colors - color maps:
# https://matplotlib.org/tutorials/colors/colormaps.html
# https://matplotlib.org/3.1.1/gallery/color/colormap_reference.html
```

```python
# sigmoid function - can work on arrays and on individual numbers
def sigmoid(t):
    s = 1.0 / (1.0 + np.exp(-t))
    return s
```

```python
# Make a 3D graph for LogisticRegression
a1 = 0.06
a2 =-0.02
b  =-0.32

x1r = np.linspace(-200, 200, 40, dtype=np.float)
x2r = np.linspace(-200, 200, 40, dtype=np.float)
y2r = np.zeros((40,40))
for i1 in range(40):
    for i2 in range(40):
        y2r[i1,i2] = sigmoid(b + a1*x1r[i1] + a2*x2r[i2])

# plot
_X1, _X2 = np.meshgrid(x1r, x2r, indexing='ij')
plt.rcParams["figure.figsize"] = (8, 8) # (width, height)
fig = plt.figure()
ax = fig.gca(projection='3d')

# mycmap = cm.YlOrBr_r
mycmap = 'jet'
ax.plot_surface(_X1, _X2, y2r, cmap=mycmap, edgecolor='gray', linewidth=0.004, antialiased=False)
ax.set_title("Logistic Regression is a Linear Model")
plt.show();
```

```python
# from Jason Brownlee
# https://machinelearningmastery.com/how-to-use-nelder-mead-optimization-in-python/
# ackley multimodal function
from numpy import arange
from numpy import exp
from numpy import sqrt
from numpy import cos
from numpy import e
from numpy import pi
from numpy import meshgrid
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# --------------------------------------------------
def objective(x, y):
    """
    # objective function
    """
    myexp1 = exp(-0.2 * sqrt(0.5 * (x**2 + y**2)))
    myexp2 = exp(0.5 * (cos(2 * pi * x) + cos(2 * pi * y)))
    val_xy = -20.0 * myexp1 - myexp2 + e + 20
    return val_xy

# --------------------------------------------------
# define range for input
r_min, r_max = -5.0, 5.0
# sample input range uniformly at 0.1 increments
xaxis = arange(r_min, r_max, 0.1)
yaxis = arange(r_min, r_max, 0.1)
# create a mesh from the axis
x, y = meshgrid(xaxis, yaxis)
# --------------------------------------------------
# compute targets
results = objective(x, y)
# --------------------------------------------------
# create a surface plot with the jet color scheme
plt.rcParams["figure.figsize"] = (12, 8)
figure = plt.figure()
axis = figure.gca(projection='3d')
axis.plot_surface(x, y, results, cmap='jet')
# show the plot
plt.show()
```
